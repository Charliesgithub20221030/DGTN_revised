```python
import numpy as np
import torch
from os.path import join as pj
import pickle

dataset = 'yoochoose1_64'

rawdir = pj('datasets' , dataset , 'raw')
```

```python
def unpickle(datadir,fn):
    with open(pj(datadir,fn),'rb') as f:
        data = pickle.load(f)
    return data
```

```python
train = unpickle(rawdir , 'train.txt')
```
